```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
#Importação e visualização dos dados

df = pd.read_excel("dados.xlsx", sheet_name="Dados_Socio")
df
```

```python
#Separação dos dados em variaveis

socios = df.iloc[:, 0]
ano = df.iloc[:, 1]
ano
```

```python
#Criação do gráfico

plt.figure(figsize=(10,5))
plt.ylabel("Número de Sócios")
plt.xlabel("Ano")
plt.title("Evolução do número de sócios do Cruzeiro")
grafico = plt.bar(ano, socios, width=0.50, color="#0000CD")
plt.bar_label(grafico, fmt="%.0f", size=10, label_type="edge")
plt.show()
```
