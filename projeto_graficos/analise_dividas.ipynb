```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
#Importação  e visualização dos dados

df = pd.read_excel("dados.xlsx", sheet_name="Dados_Dividas")
df.head()
```

```python
#Separação de cada coluna em uma variável pra usar no laço

fiscal = df.iloc[:, 0]
trab = df.iloc[:, 1]
res = df.iloc[:, 2]
total = df.iloc[:, 3]
ano = df.iloc[:, 4]
total
```

```python
#Criação do laço e dos gráficos

lista = [fiscal, trab, res, total]

def milhoes(i, pos):
    return f'R$ {i*1e-6}m'

formato = FuncFormatter(milhoes)

for i in lista:
    fig, ax = plt.subplots()
    ax.yaxis.set_major_formatter(formato)
    fig.set_size_inches(10,5)
    grafico = plt.bar(ano, i, width=0.50, color="#0000CD")
    plt.ylabel("Valor Dívida")
    plt.xlabel("Ano")
    if i.name == "Divida_Tributaria":
        plt.title("Evolução da Dívida Tributária do Cruzeiro")
    elif i.name == "Divida_Trabalhista":
        plt.title("Evolução da Dívida Trabalhista do Cruzeiro")
    elif i.name == "Total":
        plt.title("Evolução da Dívida Global do Cruzeiro")
    elif i.name == "Defict_Superavit":
        plt.title("Evuloção do Resultado do Balanço (Défict / Superavit)")
    plt.bar_label(grafico, fmt='R$%.02f', size=10, label_type="edge")
    plt.show()
```
