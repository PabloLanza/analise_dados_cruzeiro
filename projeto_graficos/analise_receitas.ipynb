```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
#Importação da base de dados

df = pd.read_excel("dados.xlsx", sheet_name="Dados_Receita")
df.head()
```

```python
#Separação de cada fonte de receita

jog = df.iloc[:, 0]
bil = df.iloc[:, 1]
socio = df.iloc[:, 2]
tv = df.iloc[:, 3]
patrocinio = df.iloc[:, 4]
out = df.iloc[:, 5]
tot = df.iloc[:, 6]
ano = df.iloc[:, 7]
```

```python
#Venda de Jogadores
def milhoes(i, pos):
    return f'R$ {i*1e-6}m'

formato = FuncFormatter(milhoes)

lista = [jog, bil, socio, tv, patrocinio, tot]


for i in lista:
    fig, ax = plt.subplots()
    ax.yaxis.set_major_formatter(formato)
    fig.set_size_inches(10,5)
    grafico = plt.bar(ano, i, width=0.50, color="#0000CD")
    if i.name == "Vendas_Jogadores":
        plt.title("Receita com vendas de jogadores")
    elif i.name == "Bilheteria_Premiacao":
        plt.title("Receita com bilheterias e premiações")
    elif i.name == "Socio_Torcedor":
        plt.title("Receita com o programa de Sócio-Torcedor")
    elif i.name == "TV":
        plt.title("Receita com TV e publicidade")
    elif i.name == "Patrocinio":
        plt.title("Receita com patrocínios")
    elif i.name == "Total":
        plt.title("Receita Total")
    plt.xlabel("Ano")
    plt.ylabel("Receita")
    plt.bar_label(grafico, fmt='R$%.02f', size=10, label_type="edge")
    plt.show()
```
